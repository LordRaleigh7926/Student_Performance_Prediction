# Imports

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
x_train = pd.read_csv("data/X_train.csv")
y_train = pd.read_csv("data/y_train.csv")
x_test = pd.read_csv("data/X_test.csv")
y_test = pd.read_csv("data/y_test.csv")

x_train.head()
```

# Preprocessing

```python
x_train = x_train.drop(["StudentID","Mjob", "Fjob", "reason", "famsup", "Dalc", "Walc"], axis=1)
x_test = x_test.drop(["StudentID","Mjob", "Fjob", "reason", "famsup", "Dalc", "Walc"], axis=1)
```

```python
y_train = y_train.drop(["StudentID"], axis=1)
y_test = y_test.drop(["StudentID"], axis=1)
```

```python
b = ["schoolsup", "paid", "activities", "nursery", "higher", "internet", "romantic"]

for i in b:
    x_train[i] = (x_train[i] == 'yes').astype(int)
    x_test[i] = (x_test[i] == 'yes').astype(int)
```

```python
x_train.head()
```

```python
def preprocess(x):
    x['school'] = (x['school'] == 'GP').astype(int)
    x['sex'] = (x['sex'] == 'M').astype(int)
    x['address'] = (x['address'] == 'U').astype(int)
    x['famsize'] = (x['famsize'] == 'GT3').astype(int)
    x['Pstatus'] = (x['Pstatus'] == 'T').astype(int)
    x["guardian"] = x.guardian.map({"mother": 1, "father": 2, "other" : 3})
```

```python
preprocess(x_train)
preprocess(x_test)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

# Models

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.svm import SVR

model2 = SVR()
model2.fit(x_train,y_train)
y_pred2 = model2.predict(x_test)
mean_squared_error(y_test, y_pred2)
```
